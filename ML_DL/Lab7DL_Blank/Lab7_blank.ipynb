# Do all pre processing on diabetes csv

```python
import pandas as pd
import numpy as np
```

# Import diabetes.csv data ser using sklearn

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

# Check the distribution of the target

```python
df['Outcome'].value_counts()
```

# Replace zeros with NaN for specific columns

```python
replace_columns = ['Glucose','SkinThickness' , 'Insulin' , 'BloodPressure' , 'BMI']
```

```python
df[replace_columns] = df[replace_columns].replace(0 , np.nan)
```

# Check for missing values

```python
df.isnull().sum()
```

# Fill missing values with median

```python
for col in replace_columns:
    median_val = df[col].median()
    df[col].fillna(median_val ,inplace=True)
```

```python
df.isnull().sum()
```

# Convert Targer data into interger code

```python
df['Outcome'] = df['Outcome'].map({'Diabetic' : 1 , 'Non Diabetic' : 0})
```

```python
df
```

# Visualize Distributions

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
sb.histplot(df)
plt.title("Distribution of Data with KDE")
plt.xlabel("Value")
plt.ylabel("Frequency/Density")
plt.grid(True, axis='y', alpha=0.5)
```

```python
sb.heatmap(df.corr() , cmap='coolwarm' , annot=True)
```

# Divide the data into input and output

```python
X = df.drop(['Outcome'] , axis = 1)
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , y , random_state=42 , test_size=0.2)
```

# import KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

# Convert all features into Z score

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_trained_scaled = scaler.fit_transform(X_train)
```

```python
x_trained_scaled
```

# Splitting the dataset into the Training set and Test set

# Create KNN Model

```python
model = KNeighborsClassifier(n_neighbors=10)
```

# Fitting KNN   on dataset

```python
model.fit(x_trained_scaled , Y_train)
```

```python
model.score(scaler.transform(X_test) , Y_test)
```

# Display classification_report

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,classification_report
```

```python
y_pred = model.predict(scaler.transform(X_test))
```

```python
cm = confusion_matrix(Y_test , y_pred)
```

```python
display = ConfusionMatrixDisplay(cm , display_labels=['Diabetic' , 'Non-Diabetic'])
```

```python
display.plot()
```

```python
print(classification_report(Y_test , y_pred))
```
