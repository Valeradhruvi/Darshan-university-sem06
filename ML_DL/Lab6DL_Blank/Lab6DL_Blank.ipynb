# Decision Tree Classifier

```python
from sklearn.datasets import load_iris 
```

# Import iris data ser using sklearn

```python
data = load_iris()
data
```

# Importing the libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

# Divide the data into input and output

```python
x = data['data']
```

```python
y = data['target']
```

```python
y
```

# Splitting the dataset into the Training set and Test set

```python
x_train , x_test , y_train ,  y_test = train_test_split(x , y , random_state=42 , test_size=0.2)
```

# Fitting DecisionTreeClassifier   on dataset

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train , y_train)
```

# Display Decision Tree

```python
from sklearn.tree import plot_tree
```

```python
%matplotlib inline
```

```python
plot_tree(model)
```

# Predict the x_test 

```python
y_pred = model.predict(x_test)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

# Import diabetes.csv dataset 

```python
df = pd.read_csv("diabetes.csv")
```

```python
df
```

# Check the distribution of the target

```python
df['Outcome'].value_counts()
```

# Replace zeros with NaN for specific columns

```python
replace_columns = ['Glucose','SkinThickness' , 'Insulin' , 'BloodPressure' , 'BMI']
```

```python
df[replace_columns] = df[replace_columns].replace(0 , np.nan)
```

# Check for missing values

```python
df.isnull().sum()
```

# Fill missing values with median

```python
for col in replace_columns:
    median_val = df[col].median()
    df[col].fillna(median_val ,inplace=True)
```

```python
df.isnull().sum()
```

# Visualize Distributions

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
sb.histplot(df)
plt.title("Distribution of Data with KDE")
plt.xlabel("Value")
plt.ylabel("Frequency/Density")
plt.grid(True, axis='y', alpha=0.5)
```

# Convert Targer data into interger code

```python
df['Outcome'] = df['Outcome'].map({'Diabetic' : 1 , 'Non Diabetic' : 0})
```

```python
df
```

```python
sb.heatmap(df.corr() , cmap='coolwarm' , annot=True)
```

# Divide the data into input and output
